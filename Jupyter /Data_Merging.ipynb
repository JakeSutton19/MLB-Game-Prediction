So this tells us that there are only two teams that need special attention: LAA and CLE because they have different name f

```python

def clean_team(Name, Team, df):
    team_df = df[df.Tm == Name].copy()
    team_df["Tm"] = Team
    team_df["Team"] = team_df["Tm"] 
    del team_df["Tm"]
    del team_df["Rdrs"]
    return team_df


#Games DF
def clean_games(Team, games_df):
    T_Games = games_df[games_df.Team == Team].copy()
    
    T_Games["y-m-d"] = pd.to_datetime(T_Games["y-m-d"])
    T_Games["Season"] = T_Games["y-m-d"].dt.year
    T_Games["Season"] = T_Games["Season"] - 1
    return T_Games

def merge_games(Team, T_Games, team_df):
    df_merge = pd.merge(T_Games, team_df, on=['Team', 'Season'])
    df_merge["champ_lev_index"] = df_merge["champ_lev_index"].fillna(0)
    df_merge = df_merge.fillna(method="ffill")
    return df_merge

```

```python
team = 'ARI'
name = 'Arizona Diamondbacks'

team_df = clean_team(name, team, df2)
t_games = clean_games(team, df1)

df_merge = merge_games(team, t_games, team_df)
df_merge
```

```python
#Team DF
file_string_1 = "data/TEAM_SEASONS/TOTAL_TEAM_SEASONS_2012-2022.csv"
file_string_2 = "data/FULL_TEAM_DATA/Full_Stats_2012-2022.csv"

df1 = pd.read_csv(file_string_1)
df2 = pd.read_csv(file_string_2) 


TEAMS_ABR = ['ARI', 'ATL', 'BAL', 'BOS', 'CHC', 'CHW', 'CIN', 'COL', 'DET', 'HOU', 'KCR', 'LAD', 'MIA', 'MIL', 
                 'MIN', 'NYM', 'NYY', 'OAK', 'PHI', 'PIT', 'SDP', 'SFG', 'SEA', 'STL', 'TBR','TEX', 'TOR', 'WSN']

TEAMS_FULL_N = [ 'Arizona Diamondbacks', 'Atlanta Braves', 'Baltimore Orioles', 'Boston Red Sox', 'Chicago Cubs', 'Chicago White Sox', 'Cincinnati Reds', 'Colorado Rockies',
       'Detroit Tigers', 'Houston Astros', 'Kansas City Royals', 'Los Angeles Dodgers', 'Miami Marlins', 'Milwaukee Brewers', 'Minnesota Twins','New York Mets', 'New York Yankees', 'Oakland Athletics',
       'Philadelphia Phillies', 'Pittsburgh Pirates', 'San Diego Padres', 'San Francisco Giants', 'Seattle Mariners', 'St. Louis Cardinals','Tampa Bay Rays', 'Texas Rangers', 'Toronto Blue Jays',
                'Washington Nationals']

#Should be missing 2 teams
len(TEAMS_ABR)
```

```python
#This will do for all teams except for CLE and LAA

all_games = []

for i in range(0, 28):
    #Assignments
    team = TEAMS_ABR[i]
    name = TEAMS_FULL_N[i]
    
    
    t_games = clean_games(team, df1)
    team_df = clean_team(name, team, df2)
    
    f_df = merge_games(team, t_games, team_df)

    all_games.append(f_df)
    print(f"{i}: {team}, {name}")

games_df = pd.concat(all_games)

```
