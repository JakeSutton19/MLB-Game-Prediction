We need to shift up all the columns so that the computer gets to learn from each game but not until after it makes its prediction

```python
games["Season_Wins"] = games["total_wins"].shift(+1)
games["Season_Loses"] = games["total_losses"].shift(+1)
games["previous_game_time"] = games["Time_Min"].shift(+1)
games["previous_runs_scored"] = games["R"].shift(+1)
games["previous_runs_allowed"] = games["RA"].shift(+1)
games["previous_innings_played"] = games["Total_Innings"].shift(+1)
games["current_streak"] = games["Streaks"].shift(+1)
games["rank"] = games["Rank"].shift(+1)
games["gb"] = games["GB"].shift(+1)
games["previous_attendance"] = games["Attendance"].shift(+1)
```

```python
from sklearn.linear_model import Ridge
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import TimeSeriesSplit

rr = Ridge(alpha=1)

split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr, n_features_to_select=20, direction="forward", cv=split, n_jobs=4)
```

```python
removed_columns = ["Outcome", "y-m-d", "GameID", "team_code", "opp_code"]
selected_columns = games.columns[~games.columns.isin(removed_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
games.loc[:,selected_columns] = scaler.fit_transform(games[selected_columns])
```

```python
sfs.fit(games[selected_columns], games["Outcome"])
```
